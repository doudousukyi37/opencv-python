## 读取图像并显示 【Read Image and Show】

```python
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

cv2.imshow('image',image)

cv2.waitKey(300) # 显示的画面计时300毫秒后再关闭
# cv2.waitKey(0)  # 取消窗口的自动关闭
```

## 图像类型与形状 【Image Type and Shape】


```python
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

print(f'cv2读取图像的数据类型为：{type(image)}')
print(f'图像数据大小为:{image.size}')

# 可通过判断图像的大小是否大于0，来判断cv2是否成功读取图像
if image.size > 0:
    print('读取图像成功')
else:
    print('读取图像失败')

height,width,channle = image.shape # 图像形状的输出顺序分别为 高度、宽度、通道数
print(f'图像的高度为:{height},图像的宽度为：{width},图像的通道数为:{channle}')
```

```python
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # 将图像转为灰度图  BGR2GRAY 表示从BGR 转 GGRAY

print(f'灰度图像的数据类型为：{type(image_gray)}')
print(f'图像数据大小为:{image_gray.size}')

height,width = image_gray.shape # 图像形状的输出顺序分别为 高度、宽度、灰度图通道数为1，因此只有两个参数
print(f'图像的高度为:{height},图像的宽度为：{width}')

cv2.imshow('image_gray',image_gray)
cv2.waitKey(300)
```

## 使用numpy生成图像 【Use Numpy to Create Image】

```python
image_numpy = np.random.randint(0,256,
                                (300,300,3),
                                dtype = np.uint8)  # 使用numpy 生成一个高300 宽300 通道数 3 每个数值随机在0到256之间的数值


cv2.imshow('image_numpy',image_numpy)
cv2.waitKey(300)
```

## 缩放图像 【Scale Image】


```python
# 将图像缩小两倍
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

height,width,channle = image.shape

new_height,new_width = int(height/2),int(width/2)

image_resize = cv2.resize(image,(new_width,new_height))

cv2.imshow('image_resize',image_resize)
cv2.waitKey(300)
```

## 裁剪图片 【Cut Image】


```python
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

x1,y1,x2,y2 = 115,130,200,200 # 在真实应用中记得判断x1,y1,x2,y2的数值是否越界，如果超过了原图的高宽可能会绘制出空的画面

image_cut = image[y1:y2,x1:x2]

cv2.imshow('image_cut',image_cut)
cv2.waitKey(300)
```

## 绘制图像 【Draw Image】

```python
image1_path = './images/IMG_6104.JPG'

image = cv2.imread(image1_path)

# 绘制矩形框
cv2.rectangle(image,
              (115,130),(200,200), # （x1,y1）,(x2,y2)
              (0,255,0),  # 矩形框颜色：B G R
              3) # 矩形框边框粗细程度

cv2.circle(image,
           (120,120), # 圆心坐标 (x,y)
           50, # 半径长度
           (0,0,255), # 红色
           3) # 圆的边框粗细程度，如果是 -1 就是填充整个圆

cv2.imshow('image_draw',image)
cv2.waitKey(300)
```
